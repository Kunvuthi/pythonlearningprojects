### Checking for missing data

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

### We can try: 
### - Add a default value for missing data
### - Delete row/column
### - Interpolation
### - Replace


```python
### Adding a default value for missing data 
df_with_0 = df.fillna(0)
df_with_0
```

```python
### or fill it with the mean
df['reviews_per_month'].mean() 
```

```python
df_with_mean = df.reviews_per_month.fillna(df['reviews_per_month'].mean()) 
df_with_mean
```

```python
### dropping NA
print(len(df.index))
```

```python
df_dropna = df.dropna()
```

```python
print(len(df_dropna.index))
```
